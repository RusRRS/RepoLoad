##### Продолжим изучать данные. Попробуем построить модель.
Посмотрим на взаимосвязи в датасете

```python
correlation_matrix = t_data.corr().round(2)
# annot = True to print the values inside the square
fig, ax = plt.subplots(figsize=(12,10)) 
sns.heatmap(data=correlation_matrix, annot=True)
```

Очевидно, что totalrooms не имеет корреляции с другими параметрами  
Уберем из тестового датасета, одновременно зададим параметры для созданиям модели.

```python
t_data.columns
```

```python
# Суть упражнения: у нас есть X - параметры и значения, y - имеющиеся значения результата. На них модель учится.
X_columns = t_data.columns[:-2] # Столбцы с 1 до предпоследнего 
y_columns = t_data.columns[-1] # Последний столбец

X_columns, y_columns
```

```python
X = t_data[X_columns]
y = t_data[y_columns]
```

```python
X
```

```python
y
```

Разделим выборку на 2 части: обучаемая и тестовая. Классическое распределение: 70/30

```python
# Разделяем выборку на 2 части. Обучаемая и тестовая.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=30)
```

```python
# Запускаем обучение модели. Модель будет называться estimator
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

estimator = LinearRegression()
# Обучение
estimator.fit(X_train, y_train)

# Предсказание
y_pred = estimator.predict(X_test)
print("R2: \t", r2_score(y_test, y_pred))
print("RMSE: \t", np.sqrt(mean_squared_error(y_test, y_pred)))
print("MAE: \t", mean_absolute_error(y_test, y_pred))
```

Оценим примерные результаты:
- **R2**:Коэффициент детерминации (0-1, показывает хватает ли наших признаков, чтобы линейная модель смогла их описать) - да
- **RMSE**:Среднеквадратичное отклонение
- **MAE**: Средняя абсолютная ошибка
